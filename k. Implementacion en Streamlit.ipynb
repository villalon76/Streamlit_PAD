### La Librería Streamlit
<br>
<img src="./Imagenes/Streamlit.png" style="width:500px;height:67px;" class="center">
<br>

**Streamlit** es un *framework* del tipo ***open source*** para la creación de aplicaciones web interactivas y basadas en datos. Está diseñado para facilitar la creación de aplicaciones de aprendizaje máquina, visualización de datos y paneles de control de manera rápida y sencilla.

Con **Streamlit** un desarrollador puede crear aplicaciones web que le permita a los usuarios interactuar con datos y modelos en tiempo real, utilizando **Python** como lenguaje de programación. Además, **Streamlit** proporciona una biblioteca de componentes preconstruidos para la visualización de datos y la creación de interfaces de usuario interactivas, lo que simplifica significativamente el proceso de desarrollo.

Una de las grandes ventajas de este *framework* es que permite crear atractivas interfaces con diseño centrado en el usuario, todo ello con pocas líneas de código. Está muy enfocado en realizar aplicaciones orientadas a datos, es por ello que tiene interrelaciones nativas con **Pandas**, así como interacciones con diferentes herramientas de visualización como **Matplotlib** y **Seaborn**.

Es posible desarrollar desde interfaces con unas cuantas líneas de código hasta gráficos, botones o deslizadores, entre otras funcionalidades propias de su naturaleza para con ello obtener aplicaciones interactivas y sencillas de utilizar.

**Streamlit** no es la única librería para la creación de aplicaciones web interactivas y basadas en datos, existen diversas alternativas. Sin embargo es una de las librerías de mayor uso actualmente, pero algunas de las opciones adicionales son:

* [Voilá](https://github.com/voila-dashboards/voila).
* [Django](https://www.djangoproject.com).
* [Pyramid](https://trypyramid.com).
* [Flask](https://flask.palletsprojects.com/en/2.2.x).
* [Web2py](http://www.web2py.com).
* [TurboGears](https://turbogears.org).
* [Tkinter](https://docs.python.org/es/3/library/tk.html).

2. El ***Gráfico de Línea*** de las ganancias para cada una de las personas en el **DataFrame** con el objetivo de visualizar la manera como se distribuyeron sus percepciones a lo largo del año. ***Se pretende que el usuario pueda cambiar interactivamente a la persona.***


3. La ***Matriz de Correlación*** de los meses presentes en el **DataFrame** con el objetivo de mostrar la relación existente entre cada uno de los meses. ***Se pretende que el usuario pueda modificar interactivamente los meses de la matriz de correlación, así como el mapa de colores que se está empleando.***

Para ello, el **Panel de Configuración** debe incluir los controles para cada una de las acciones interactivas que el usuario podrá modificar. Los elementos del **Panel de Configuración** se muestran con métodos similares a los que fueron descritos con anterioridad, pero se les agrega el método `.sidebar` para indicar que estarán contenidos dentro del **Dashboard**. Los 5 elementos (indicados en la figura previa) son: 

1. Se muestra una imagen por medio del método `.sidebar.image()`, así como un pequeño texto a manera de título por medio del método `.sidebar.markdow()`.


2. Selección del mes para el ***Histograma***. Primeramente se define una lista con las opciones (los meses, son 12 para este **DataFrame**) a emplear, se define un valor predeterminado (*default*), y finalmente se indica un selector de valor único por medio del método `.sidebar.selectbox()`.  


3. Selección de las personas para el ***Gráfico de Línea***. Primeramente se define una lista con las opciones (las personas, son 10 para este **DataFrame**) a emplear, se define un valor predeterminado (*default*), y finalmente se indica un selector de valor único por medio del método `.sidebar.selectbox()`.


4. Selección del mapa de color para el ***Histograma***. Primeramente se define una lista con las opciones (los mapas de color, solamente se definen 10 de ellos) a emplear, se define un valor predeterminado (*default*), y finalmente se indica un selector de valor único por medio del método `.sidebar.selectbox()`.


5. Selección de los meses para el ***Histograma***. Se indica un selector de valor múltiple por medio del método `.sidebar.multiselect()`, indicando en sus parámetros la lista con las opciones a incluir en el ***Histograma*** (los meses, son 12 para este **DataFrame**), así como el valor predeterminado (*default*).

La descripción detallada del método `.selectbox()` se puede consultar [en esta liga](https://docs.streamlit.io/library/api-reference/widgets/st.selectbox).

La descripción detallada del método `.multiselect()` se puede consultar [en esta liga](https://docs.streamlit.io/library/api-reference/widgets/st.multiselect).

````python
#Renderizado de la Imagen y el Título en el Dashboard
st.sidebar.image(Logo, width = 200)
st.sidebar.markdown("## CONFIGURACIÓN")

#Selector del Mes en el Dashboard
vars_mes = ['ENE','FEB','MAR','ABR','MAY','JUN','JUL','AGO','SEP','OCT','NOV','DIC']
default_hist = vars_mes.index('ENE')
histo_selected = st.sidebar.selectbox('Mes para el Histograma:', vars_mes, index = default_hist)

#Selector de las Personas en el Dashboard
vars_per = ['Iñaki González','María Cázares','José García','Jérémie Muñoz','Agnès Villalón','Bérénice Pitkämäki','Geneviève Rukajärvi','Hélène Ñuñoz','Ñaguí Grönholm','Iván Földváry']
default_pers = vars_per.index('Iñaki González')
ganan_selected = st.sidebar.selectbox('Ganancias Personales:', vars_per, index = default_pers)

#Selector del Mapa de Color en el Dashboard
vars_cmap = ['crest', 'viridis', 'OrRd','Pastel1', 'Paired', 'ocean', 'jet', 'Accent','Dark2','twilight']
default_cmap = vars_cmap.index('crest')
color_selected = st.sidebar.selectbox('Color de la Matriz de Correlación:', vars_cmap, index = default_cmap)

#Selector de los Meses para el Histograma en el Dashboard
mes_multi_selected = st.sidebar.multiselect('Elementos de la Matriz de Correlación:', vars_mes, default = vars_mes)
````

### Inicialización de los Elementos en la Sección de Contenidos

Adicional a los tres elementos que se agregaron inicialmente en la **Sección de Contenidos** (imagen, título, descripción y **DataFrame** a ser empleado), se realizará la conexión entre los elementos del **Panel de Configuración** y los elementos de la **Sección de Contenidos** describiendo cada uno de ellos.

El primer elemento corresponde al ***Histograma*** de las columnas correspondientes a los meses con el objetivo de visualizar gráficamente la distribución de frecuencia de las ganancias de todos los clientes cada mes. ***Se pretende que el usuario pueda cambiar interactivamente el mes.*** Para ello, se realizarán las siguientes operaciones:

1. Se definirán dos columnas en la **Sección de Contenidos** de la página web con la finalidad de mostrar un gráfico en cada una de ellas. Esto se realiza por medio del método `.columns()` de **Streamlit**.


2. Se indicará un título para el gráfico a ser presentado por medio del método `.subheader()` aplicado a la columna correspondiente. ***Nota:*** En el caso de que no se estén utilizando columnas en la **Sección de Contenidos**, el método `colum_izq.subheader()` se reemplaza por `st.subheader()`.


3. Se inicializa el gráfico por medio del método `.subplots()` de **Matplotlib**, el cual regresa el gráfico así como el encapsulamiento de los elementos del gráfico.


4. Se genera el gráfico por medio del método `.histplot()` de **Seaborn** con la información del **DataFrame**, extrayendo del **DataFrame** la información del mes que haya sido seleccionado por el usuario en el **Panel de Configuración** a través de `histo_selected`. En este punto se agregan elementos adicionales como el estilo del gráfico, así como las etiquetas de los ejes y el título.


5. Finalmente se renderiza la imagen en la **Sección de Contenidos** de la página web por medio del método `.pyplot()` de **Streamlit**. ***Nota:*** En el caso de que no se estén utilizando columnas en la **Sección de Contenidos**, el método `colum_izq.pyplot()` se reemplaza por `st.pyplot()`.

La descripción detallada del método `.pyplot()` se puede consultar [en esta liga](https://docs.streamlit.io/library/api-reference/charts/st.pyplot).

````python
#Definición de las columnas
colum_izq, colum_der = st.columns(2)

#Título para el gráfico
colum_izq.subheader('Histograma')

#Inicialización del gráfico
fig1, ax1 = plt.subplots()

#Generación del gráfico
sns.set(style = "darkgrid")
sns.histplot(data = datos_df[histo_selected])
ax1.set_title('Histograma de Valores')
ax1.set_xlabel(histo_selected)
ax1.set_ylabel('Frecuencia')

#Renderización del gráfico
colum_izq.pyplot(fig1)
````

El segundo elemento corresponde al ***Gráfico de Línea*** de las ganancias para cada una de las personas en el **DataFrame** con el objetivo de visualizar la manera como se distribuyeron sus percepciones a lo largo del año. ***Se pretende que el usuario pueda cambiar interactivamente a la persona.*** Para ello, se realizarán las siguientes operaciones:

1. Se indicará un título para el gráfico a ser presentado por medio del método `.subheader()` aplicado a la columna correspondiente. ***Nota:*** En el caso de que no se estén utilizando columnas en la **Sección de Contenidos**, el método `colum_der.subheader()` se reemplaza por `st.subheader()`.


2. Se inicializa el gráfico por medio del método `.subplots()` de **Matplotlib**, el cual regresa el gráfico así como el encapsulamiento de los elementos del gráfico.


3. Se genera el gráfico por medio del método `.plot()` de **Matplotlib** con la información del **DataFrame**, pero para ello se realizará el proceso de generación del gráfico que se revisó en subtemas previos, extrayendo del **DataFrame** la información de la persona que haya sido seleccionada por el usuario en el **Panel de Configuración** a través de `ganan_selected`. En este punto se agregan elementos adicionales como las etiquetas de los ejes y el título.


4. Finalmente se renderiza la imagen en la **Sección de Contenidos** de la página web por medio del método `.pyplot()` de **Streamlit**. ***Nota:*** En el caso de que no se estén utilizando columnas en la **Sección de Contenidos**, el método `colum_der.pyplot()` se reemplaza por `st.pyplot()`.

````python
#Título para el gráfico
colum_der.subheader('Ganancias')

#Inicialización del gráfico
fig2, ax2 = plt.subplots()

#Generación del gráfico
if ganan_selected == 'Iñaki González':
    periodo_df = datos_df.iloc[0]
elif ganan_selected == 'María Cázares':
    periodo_df = datos_df.iloc[1]
elif ganan_selected == 'José García':
    periodo_df = datos_df.iloc[2]
elif ganan_selected == 'Jérémie Muñoz':
    periodo_df = datos_df.iloc[3]
elif ganan_selected == 'Agnès Villalón':
    periodo_df = datos_df.iloc[4]
elif ganan_selected == 'Bérénice Pitkämäki':
    periodo_df = datos_df.iloc[5]
elif ganan_selected == 'Geneviève Rukajärvi':
    periodo_df = datos_df.iloc[6]
elif ganan_selected == 'Hélène Ñuñoz':
    periodo_df = datos_df.iloc[7]
elif ganan_selected == 'Ñaguí Grönholm':
    periodo_df = datos_df.iloc[8]
elif ganan_selected == 'Iván Földváry':
    periodo_df = datos_df.iloc[9]
else:
    periodo_df = datos_df
periodo_df = periodo_df.transpose()
periodo_df = periodo_df.to_frame()
periodo_df = periodo_df.rename(columns = {1: 'MES'})
periodo_df = periodo_df.drop(['NOMBRE','APELLIDO','CIUDAD'])
plt.plot(periodo_df)
ax2.set_title('Ganancias Mensuales por Persona')
ax2.set_xlabel(ganan_selected)
ax2.set_ylabel('Ganancias')

#Renderización del gráfico
colum_der.pyplot(fig2)
````

El tercer y último elemento corresponde a la ***Matriz de Correlación*** de los meses presentes en el **DataFrame** con el objetivo de mostrar la relación existente entre cada uno de los meses. ***Se pretende que el usuario pueda modificar interactivamente los meses de la matriz de correlación, así como el mapa de colores que se está empleando.*** Para ello, se realizarán las siguientes operaciones:

1. Se indicará un título para el gráfico a ser presentado por medio del método `.subheader()`. 


2. Se inicializa el gráfico por medio del método `.subplots()` de **Matplotlib**, el cual regresa el gráfico así como el encapsulamiento de los elementos del gráfico.


3. Se determina la ***Matriz de Correlación*** por medio del método `.corr()` de **Pandas**, aplicado a las variables contenidas en la lista definida por el usuario en el **Panel de Configuración** a través de `mes_multi_selected`, para posteriormente generar el gráfico por medio del método `.heatmap()` de **Seaborn**. Adicionalmente, se emplea la selección del usuario en el **Panel de Configuración** del mapa de color a ser empleado para el gráfico por medio de `color_selected`.


4. Finalmente se renderiza la imagen en la **Sección de Contenidos** de la página web por medio del método `.pyplot()` de **Streamlit**. 

````python
#Título para el gráfico
st.subheader('Matriz de Correlación')

#Inicialización del gráfico
fig3, ax3 = plt.subplots()

#Generación del gráfico
df_corr = datos_df[mes_multi_selected].corr()
sns.heatmap(df_corr, annot = True, fmt='.2f', cmap = color_selected)

#Renderización del gráfico
st.pyplot(fig3)
````

````python
# =================================================================
# == INSTITUTO TECNOLOGICO Y DE ESTUDIOS SUPERIORES DE OCCIDENTE ==
# ==          ITESO, UNIVERSIDAD JESUITA DE GUADALAJARA          ==
# ==                                                             ==
# ==             MAESTRÍA EN SISTEMAS COMPUTACIONALES            ==
# ==              PROGRAMACIÓN PARA ANÁLISIS DE DATOS            ==
# ==                  IMPLEMENTACIÓN EN STREAMLIT                ==
# =================================================================

#Importación de librerías
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import streamlit as st
from skimage import io

#Lectura de los datos de los Administradores desde el archivo CSV
datos_df = pd.read_csv('./Datos/Datos_DF.csv')

#Lectura de la Imagen
Logo = io.imread("./Imagenes/ITESO_Logo.png")

#Renderizado de la imagen
st.image(Logo, width = 500)

#Renderizado del texto
st.title("Uso Básico de Streamlit")
st.subheader(":blue[Streamlit es un *framework* para la creación de aplicaciones web interactivas y basadas en datos.]")

#Renderizado del texto
st.markdown("**Manejo de Información de Clientes**")
st.markdown(":blue[Este **DataFrame** contiene información de varias personas, las ciudades donde viven, así como sus ganancias a lo largo de un año.]")

#Renderizado del DataFrame
st.dataframe(datos_df)

#Renderizado de la Imagen y el Título en el Dashboard
st.sidebar.image(Logo, width = 200)
st.sidebar.markdown("## CONFIGURACIÓN")

#Selector del Mes en el Dashboard
vars_mes = ['ENE','FEB','MAR','ABR','MAY','JUN','JUL','AGO','SEP','OCT','NOV','DIC']
default_hist = vars_mes.index('ENE')
histo_selected = st.sidebar.selectbox('Mes para el Histograma:', vars_mes, index = default_hist)

#Selector de las Personas en el Dashboard
vars_per = ['Iñaki González','María Cázares','José García','Jérémie Muñoz','Agnès Villalón','Bérénice Pitkämäki','Geneviève Rukajärvi','Hélène Ñuñoz','Ñaguí Grönholm','Iván Földváry']
default_pers = vars_per.index('Iñaki González')
ganan_selected = st.sidebar.selectbox('Ganancias Personales:', vars_per, index = default_pers)

#Selector del Mapa de Color en el Dashboard
vars_cmap = ['crest', 'viridis', 'OrRd','Pastel1', 'Paired', 'ocean', 'jet', 'Accent','Dark2','twilight']
default_cmap = vars_cmap.index('crest')
color_selected = st.sidebar.selectbox('Color de la Matriz de Correlación:', vars_cmap, index = default_cmap)

#Selector de los Meses para el Histograma en el Dashboard
mes_multi_selected = st.sidebar.multiselect('Elementos de la Matriz de Correlación:', vars_mes, default = vars_mes)

#Definición de las columnas
colum_izq, colum_der = st.columns(2)

#Título para el gráfico
colum_izq.subheader('Histograma')

#Inicialización del gráfico
fig1, ax1 = plt.subplots()

#Generación del gráfico
sns.set(style = "darkgrid")
sns.histplot(data = datos_df[histo_selected])
ax1.set_title('Histograma de Valores')
ax1.set_xlabel(histo_selected)
ax1.set_ylabel('Frecuencia')

#Renderización del gráfico
colum_izq.pyplot(fig1)

#Título para el gráfico
colum_der.subheader('Ganancias')

#Inicialización del gráfico
fig2, ax2 = plt.subplots()

#Generación del gráfico
if ganan_selected == 'Iñaki González':
    periodo_df = datos_df.iloc[0]
elif ganan_selected == 'María Cázares':
    periodo_df = datos_df.iloc[1]
elif ganan_selected == 'José García':
    periodo_df = datos_df.iloc[2]
elif ganan_selected == 'Jérémie Muñoz':
    periodo_df = datos_df.iloc[3]
elif ganan_selected == 'Agnès Villalón':
    periodo_df = datos_df.iloc[4]
elif ganan_selected == 'Bérénice Pitkämäki':
    periodo_df = datos_df.iloc[5]
elif ganan_selected == 'Geneviève Rukajärvi':
    periodo_df = datos_df.iloc[6]
elif ganan_selected == 'Hélène Ñuñoz':
    periodo_df = datos_df.iloc[7]
elif ganan_selected == 'Ñaguí Grönholm':
    periodo_df = datos_df.iloc[8]
elif ganan_selected == 'Iván Földváry':
    periodo_df = datos_df.iloc[9]
else:
    periodo_df = datos_df
periodo_df = periodo_df.transpose()
periodo_df = periodo_df.to_frame()
periodo_df = periodo_df.rename(columns = {1: 'MES'})
periodo_df = periodo_df.drop(['NOMBRE','APELLIDO','CIUDAD'])
plt.plot(periodo_df)
ax2.set_title('Ganancias Mensuales por Persona')
ax2.set_xlabel(ganan_selected)
ax2.set_ylabel('Ganancias')

#Renderización del gráfico
colum_der.pyplot(fig2)

#Título para el gráfico
st.subheader('Matriz de Correlación')

#Inicialización del gráfico
fig3, ax3 = plt.subplots()

#Generación del gráfico
df_corr = datos_df[mes_multi_selected].corr()
sns.heatmap(df_corr, annot = True, fmt='.2f', cmap = color_selected)

#Renderización del gráfico
st.pyplot(fig3)
````

### Despliegue en un Sitio Web

**Streamlit** permite el despliegue (***deployment***) del sitio web diseñado a partir de los datos y su procesamiento por medio de la librería **Pandas** de **Python**, el cual puede realizarse de manera pública o de manera privada.

Una de las opciones es a través de un repositorio de **GitHub** (cuyo acceso es a través de [esta liga](https://github.com)) en el cual se deberá de incluir el archivo `Streamlit_PAD.py`, así como las carpetas `Datos` e `Imagenes`. Adicionalmente, se debe incluir un archivo `requirements.txt` el cual indica las versiones de las librerías que se estarán ejecutando, el contenido de este archivo puede ser como se muestra a continuación:
